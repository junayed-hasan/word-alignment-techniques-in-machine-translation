## Comparative analysis of performance with varying threshold values

```python
import numpy as np
from scipy.interpolate import make_interp_spline

# Creating smooth curves by interpolating the data
threshold_smooth = np.linspace(min(threshold), max(threshold), 300)

# Interpolating each curve
precision_smooth = make_interp_spline(threshold, precision)(threshold_smooth)
recall_smooth = make_interp_spline(threshold, recall)(threshold_smooth)
aer_smooth = make_interp_spline(threshold, aer)(threshold_smooth)

# Plotting the smoothed graph
plt.figure(figsize=(10, 6))

# Plot each smoothed curve
plt.plot(threshold_smooth, precision_smooth, label="Precision", linewidth=2)
plt.plot(threshold_smooth, recall_smooth, label="Recall", linewidth=2)
plt.plot(threshold_smooth, aer_smooth, label="AER", linewidth=2)

# Add labels and title
plt.xlabel("Threshold", fontsize=14)
plt.ylabel("Values", fontsize=14)
plt.title("Variation of Word Alignment Performance with Threshold for 10000 Sentences", fontsize=16)

# Add grid, legend, and fine-tune layout
plt.grid(True, which='both', linestyle='--', linewidth=0.7)
plt.legend(loc="best", fontsize=12)
plt.xticks(threshold)

# Display the plot
plt.tight_layout()
plt.show()
```
